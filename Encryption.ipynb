```
import cv2
import string
 
enc = {} 
dec = {} 

for i in range(255):
    enc[chr(i)] = i    
    dec[i] = chr(i)

pic = cv2.imread("pic.jpg")

i = pic.shape[0]
j = pic.shape[0]

key = input("Enter the Secret Key:" )
text = input("Enter the text to encrypt: ")

kl = 0  #length of key

z=0; x=0; y=0

l = len(text)

#Encryption, the code will hide data in top left corer

for i in range (l):
    pic[x, y, z] = enc[text[i]] ^ enc[key[kl]]
    x = x+1
    y = y+1
    z = (x+1) % 3 
    kl = (kl+1) % len(key)

cv2.imwrite("e_pic2.jpg", pic)
b = cv2.imread("e_pic2.jpg")
import matplotlib.pyplot as plt
plt.imshow(b)
plt.show()
print("!Data hidden successfully!")
```
